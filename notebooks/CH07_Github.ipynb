### [코드 예제 7-2: 데이터 시리즈로 만들 리스트 만들기]

```python
import random

list_korean = [random.randint(50, 100) for x in range(10)] #국어 시험 성적
list_math = [random.randint(50, 100) for x in range(10)] #수학 시험 성적

print("10명의 학생의 국어 점수 {} \n10명의 학생의 수학 점수 {}".format(list_korean, list_math))
```

### [코드 예제 7-3: 리스트를 데이터 시리즈로 바꾸기]

### [코드 예제 7-56-2: 판다스를 이용한 데이터 정보 보기] 

```python
print(df['강수량'].count()) # 카운트 
print(df['강수량'].mean()) # 평균 
print(df['강수량'].sum()) # 합 
print(df['강수량'].std()) # 표준편차
print(df['강수량'].min()) # 최소값
print(df['강수량'].max()) # 최대값 
print(df['강수량'].median()) # 중앙값 
print(df['강수량'].quantile(.25)) # 25%위치한 값 
print(df['강수량'].quantile(.75)) # 75%에 위치한 값 
```

### [코드 예제 7-57: 데이터프레임의 정보를 사용하여 정규화하기]

```python
df['평균기온z값'] = (df['평균기온'] - df['평균기온'].mean()) / df['평균기온'].std()
df.head()
```

### [코드 예제 7-58: 여러 정보 구하기]

```python
# 평균과 표준편차 구하기 
print(df['강수량'].agg(['mean', 'std'])) 
# 평균과 표준편차 인덱싱
print(df['강수량'].agg(['mean', 'std'])[0])
print(df['강수량'].agg(['mean', 'std'])[1])
```

### [코드 예제 7-62: 날짜 데이터를 명목변수로 바꾸기] 

```python
# 날짜 정보를 활용하기 위한 모듈 불러오기 
import datetime as dt 

# 날짜 형식으로 변경  
df['날짜'] = pd.to_datetime(df['날짜']) 

# 날짜 데이터에서 필요한 값 추출 
df["년"] = df["날짜"].dt.year #연도 추출 
df["월"] = df["날짜"].dt.month #해당 달 추출 
df["요일"] = df["날짜"].dt.dayofweek #요일 추출 

df.info()
```
