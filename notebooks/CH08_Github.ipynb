### [코드 예제 8-1: 실습에 필요한 코드 불러오기] 

```python
# 필요한 라이브러리 부르기 
import pandas as pd

# 데이터 시각화에 필요한 라이브러리들 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

### [코드 예제 8-2: 선 그래프 그리기] 

```python
# 데이터 만들기 
x = [x for x in range(1, 11)] # 1부터 10까지의 리스트 
y = [y**3 for y in range(1, 11)] # 1부 10까지의 3자승 값을 저장한 리스트 

# 선 그래프 그리기
sns.lineplot(x=x, y=y) 
```

### [코드 예제 8-3: 그래프에 제목 붙이기] 

```python
sns.lineplot(x=x, y=y) # 각 좌표를 잇는 선그래프 
sns.scatterplot(x, y) # 각 좌표에 점을 표시하는 산점도 그리기 

plt.title('Test') #이미지 제목 설정
plt.xlabel('X-axis') #x축 레이블 설정
```

### [코드 예제 8-4: 데이터프레임을 활용해 시각화하기] 

```python
# 데이터 프레임 만들기 
df0 = pd.DataFrame({'x':x, 'y':y}) 

#데이터 프레임을 사용하여 라인 그래프 그리기
sns.lineplot(x='x', y='y', data=df0) 
plt.title('Test Plot')
```

### [코드 예제 8-5: 두 선그래프 그리기] 

```python
# 데이터 프레임 만들기 
df0 = pd.DataFrame({'x':x, 'y':y}) 
df1 = pd.DataFrame({"x":[x for x in range(1, 11)], 
                    "y":[x**2 for x in range(1, 11)]})

# 선 그래프 그리기 
sns.lineplot(data=df0, x='x', y='y') 
sns.lineplot(data=df1, x='x', y='y') 

# 정보 표시 
plt.title('title') # 제목
plt.legend(['power**3', 'power**2']) # 범례
```

### [코드 예제 8-15: 박스플롯 그리기] 

```python
# 캔버스 설정 
figure, axs = plt.subplots(ncols=2) #
figure.set_size_inches(10,5) 

# 박스플롯
ax0 = sns.boxplot(x=df["total_bill"], ax=axs[0]) # 첫번째 이미지 
ax1 = sns.boxplot(y=df["total_bill"], ax=axs[1]) # 두번째 이미지 

#이미지 정보 표시
ax0.set(title='Box Plot', xlabel='X Axis: Total_bill') 
ax1.set(title='Box Plot', ylabel='Y Axis: Total_bill') 
```

### [코드 예제 8-16: 박스플롯에 표시되는 정보 계산]

```python
# Q1, Median, Q3 값
a,b,c = df['total_bill'].quantile([.25, .5, .75])
iqr = c - a # IQR 계산 
min = a-1.5*iqr # 최소값
max = c+1.5*iqr # 최대값 
out1 = len(df[df['total_bill']<min]) # 이상치 
out2 = len(df[df['total_bill']>max]) # 이상치 

print('Total Bill:\n중앙값: {0:8.3f},\nQ1: {1:8.3f}\nQ3: {2:8.3f}\nIQR: {3:8.3f}\n' \
      'Minimum: {4:8.3f}\nMaximum: {5:8.3f}\nMinimum보다 작은 이상치: {6}개\nMaximum보다 큰 이상치: {7}개'\
      .format(b, a, c, iqr, min, max, out1, out2))
```

### [코드 예제 8-25: 피벗테이블 만들기]

```python
# 피벗 테이블 만들기 
bike = bike_raw.pivot('day_of_week','hour','bike_rent') 
print(bike)
```

### [코드 예제 8-26: 피벗테이블을 사용하여 히트맵 시각화하기] 

```python
# 이미지 설정 
plt.figure(figsize=(5,5)) 

# 히트맵 만들기 
ax = sns.heatmap(bike) 

# 이미지 정보 표시 
ax.set(title="Bike Demand Heatmap", # 제목 
       ylabel='Day of Week', xlabel="Hour", # x축 y축의 제목 표시 
       yticklabels=['Mon','Tue','Wed','Thr','Fri','Sat','Sun']) # y축 요일명 변경 

plt.show()
```

### [코드 예제 8-27: 히트맵의 색 패턴 변경]

```python
# 캔버스 설정 
fig = plt.figure() 
fig.set_size_inches(10,10) 

# 히트맵 색 변경을 위하여 반복문 설정 
for i, colors in enumerate(['PuRd','Greys','Reds','PuBuGn']): # 표시할 색 
  # 이미지 위치 설정: 2x2이미지 중 왼쪽위에서 부터 i+1번째에 이미지 출력 
  axs = fig.add_subplot(2,2,i+1) 
  sns.heatmap(bike, ax=axs, cmap=colors).set(title="cmap: {}".format(colors))
fig.show()
```

### [코드 예제 8.6.2-1: 데이터프레임의 데이터를 문자열로 바꾸기] 

```python
# 데이터프레임을 하나의 문자열로 만들기 
string_of_keywords = df['keywords'].sum() 

# 첫 30자 표기 
print(string_of_keywords[0:30]) 
```

### [코드 예제 8-33: .split()를 이용한 리스트 만들기]

### [코드 예제 8-34: 단어:빈도 형태의 딕셔너리 만들기] 

```python
# 필요한 함수 블러오기 
from collections import Counter 

# 단어:빈도의 딕셔너리 만들기 
counts = Counter(list_of_keywords)
print(counts)
```

### [코드 예제 8-36: 워드클라우드 시각화에 필요한 모듈 불러오기] 

```python
# 필요한 모듈 불러오기
from wordcloud import WordCloud
import matplotlib.pyplot as plt 
```

### [코드 예제 8-37: 워드클라우드 만들기] 

```python
# 워드클라우드 객체 선언 및 사용할 폰트 지정
wc1 = WordCloud(font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf", 
               background_color="white", max_font_size=60) 

# 단어빈도수를 활용하여 워드클라우드 만들기 
cloud1 = wc1.generate_from_frequencies(dict(counts)) 
```

### [코드 예제 8-38: 워드 클라우드 시각화] 

```python
# 시각화 이미지 크기 설정 
plt.figure(figsize=(7.5, 5))
plt.axis('off') # 워드클라우드의 경계선 없애기 
plt.imshow(cloud1) # 시각화할 워드클라우드 객체 
plt.show() # 이미지 표시
```

### [코드 예제 8-39: 워드클라우드 배경색 변경하기] 

```python
# 배경색을 변경하여 워드클라우드 시각화 
wc2 = WordCloud(font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf", 
               background_color="black", max_font_size=60) 
cloud2 = wc2.generate_from_frequencies(dict(counts))

# 여러 개의 워드클라우드 시각화 
# 캔버스 설정 
fig = plt.figure()  
fig.set_size_inches(15,10) # 캔버스의 크기 조정 

 # 반복문을 사용해 워드클라우드 시각화 
for i, wc in enumerate([cloud1, cloud2]):
  ax1 = fig.add_subplot(1,2,i+1)
  ax1.axis('off') 
  ax1.imshow(wc)   

#시각화 
fig.show()
```

### [코드 예제 8-40: max_font_size 아규먼트를 변경하여 워드 클라우드 시각화] 

```python
# 캔버스 설정 
fig = plt.figure() 
fig.set_size_inches(15,10) 

# 총 네 개의 워드 클라우드 생성 
for i in range(4):  
  
  # 매번 max_font_size를 달리하여 새로운 워드클라우드 이미지 만들기 
  wc = WordCloud(font_path="/usr/share/fonts/truetype/nanum/NanumGothic.ttf",
               background_color="white", max_font_size=120/(i+1)) 
  cloud = wc.generate_from_frequencies(dict(counts))

  ax1 = fig.add_subplot(2,2,i+1) #새로 만든 이미지를 i+1 번 인덱스에 배치 

  ax1.axis('off') 
  ax1.imshow(cloud) 

  # 제목 변경 
  ax1.set(title='Max Font Size= {}'.format(120/(i+1))) 

fig.show()
```
